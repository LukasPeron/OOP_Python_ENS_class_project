# Fichier démonstration package Feynman_package #

Ce package a pour but d'offrir la possibilité de créer et de manipuler des diagrammes de Feynman en langage python. Le package comporte pour l'instant un module complet pour la QED. Il permet de créer un diagramme, de le modifier, de l'afficher ainsi que d'en afficher l'intégrale associée.

### Exemple 1 : Diffusion Compton ###

```python
from My_package.feynman_package import *

#création du diagramme vide
d = Diagram(channel="t")
#création des vertexs
A = QED_Vertex("A", d)
B = QED_Vertex("B", d)
#création des particules
e1 = Electron("e1", 'p')
e2 = Electron("e2", 's')
e3 = Electron("e3", 'q')
gamma1 = Photon("gamma1", 'k')
gamma2 = Photon("gamma2", 'w')
#connexion particules/vertexs
A.connect_particle_in(e1, gamma1)
A.connect_vertex_out(B, e2)
B.connect_particle_out(e3, gamma2)
#ajout des vertexs remplis au diagramme
d.add_vertex(A, B)
#affichage
d.display('horizontal=A to B', save_files=True, show_code=True, show_image=True)
d.integral(show_code=True)
```

### Exemple 2 : Diagramme avec boucle ###

```python
from My_package.feynman_package import *

d = Diagram(channel="t")

A = QED_Vertex('A', d)
B = QED_Vertex('B', d)
C = QED_Vertex('C', d)
D = QED_Vertex('D', d)

e1 = AntiElectron('e1', 'p_1')
e2 = Electron('e2', 'p_2')
e3 = Electron('e3', 'e_3')
e4 = AntiElectron('e4', 'e_4')
e5 = Electron('e5', 'p_5')
e6 = AntiElectron('e6', 'p_6')
gamma1 = Photon('gamma1', 'k_1')
gamma2 = Photon('gamma2', 'k_2')

A.connect_particle_in(e2, e1)
A.connect_vertex_out(B, gamma1)
B.connect_loop_out(C, e3, e4)
C.connect_vertex_out(D, gamma2)
D.connect_particle_out(e5, e6)
d.add_vertex(A, B, C, D)
d.display('horizontal=A to B')
d.integral()
```

### Exemple 3 : Diffusion par le cannal s ###

```python
from My_package.feynman_package import *

d = Diagram(channel="s")

A = QED_Vertex("A", d)
B = QED_Vertex("B", d)
C = QED_Vertex("C", d)
D = QED_Vertex("D", d)

e1 = Electron("e1", "p")
e2 = Electron("e2", "q")
e3 = Electron("e3", "p'")
e4 = Electron("e4", "q'")
w1 = Electron("w1", "w_1")
w2 = AntiElectron("w2", "w_2")
gamma1 = Photon("gamma1", "p' - p")
gamma2 = Photon("gamma2", "p' - p")

A.connect_particle_in(e1)
A.connect_particle_out(e3)
A.connect_vertex_out(B, gamma1)
B.connect_loop_out(C, w1, w2)
C.connect_vertex_out(D, gamma2)
D.connect_particle_in(e4)
D.connect_particle_out(e2)

d.add_vertex(A, B, C, D)
d.display("vertical= A to B")
d.integral()
```
